# Quickstart

This walks through a minimal `a3dasm` run end to end. The system is a graph of agents: a hub *strategizer* delegates to specialists (literature reviewer, data generator, implementer, critic), keeps a Popperian hypothesis ledger, and produces a reproducible `pipeline.ipynb` that must pass a reproduction gate before the run is considered closed.

The only input you must provide is a `PROBLEM_STATEMENT.md` in the study directory.

## Install

```bash
pip install "a3dasm @ git+ssh://git@github.com/bessagroup/a3dasm.git@v0.1.0"
```

`a3dasm` depends on [f3dasm](https://github.com/bessagroup/f3dasm) for the data-driven primitives (`ExperimentData`, `Domain`, `DataGenerator`, the `Pipeline`).

## 1. Describe the problem

A study is just a directory. At minimum it holds a `PROBLEM_STATEMENT.md`. Everything else (the domain, the sampling, the surrogate, the optimisation) is decided by the agents.

````python
from pathlib import Path

study_dir = Path("studies/quickstart")
study_dir.mkdir(parents=True, exist_ok=True)

(study_dir / "PROBLEM_STATEMENT.md").write_text(
    "Minimise the 2D Branin function over its standard domain.\n"
    "Report the best design found and the objective value there.\n"
)
````

## 2. Run the agent loop

`AgenticRun` reads the problem statement, builds the default graph, and runs to a gated deliverable. Pick a model your configured backend can serve (here, Claude Haiku). `execute()` returns the final report text.

````python
from a3dasm import AgenticRun

report = AgenticRun(
    study_dir=study_dir,
    model="claude-haiku-4-5-20251001",
    eval_budget=200,
).execute()

print(report)
````

## 3. Read the deliverable

The run writes `pipeline.ipynb` in the study directory. That notebook *is* the deliverable: its markdown cells hold the writeup and its code cells reproduce the headline result from the canonical evaluation ledger. The reproduction gate has already executed it end to end, so it runs as-is.

Run artifacts (retrospectives, diagnostics, critic reviews, the delegation transcripts) live under `studies/quickstart/runs/<timestamp>/debug/`.
